# Nettoyage des données pour les poissons

```python
import zipfile
import pandas as pd

zip_path = r"Seine_donnees.zip"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    # Charger chaque CSV dans une variable dédiée
    indic = pd.read_csv(zip_ref.open("Poissons/poissons_indicateurs.csv"))
    obs = pd.read_csv(zip_ref.open("Poissons/poissons_observations.csv"))
    ops = pd.read_csv(zip_ref.open("Poissons/poissons_operations.csv"))
    stats = pd.read_csv(zip_ref.open("Poissons/poissons_stations.csv"))

    qual_analyse = pd.read_csv(zip_ref.open("Qualite des Cours d'Eau/qualite_eau_analyse_pc.csv"))
    qual_cond_env = pd.read_csv(zip_ref.open("Qualite des Cours d'Eau/qualite_eau_condition_env_pc.csv"))
    qual_ops = pd.read_csv(zip_ref.open("Qualite des Cours d'Eau/qualite_eau_operation_pc.csv"))
    qual_stats = pd.read_csv(zip_ref.open("Qualite des Cours d'Eau/qualite_eau_station_pc.csv"))

# Affichage de la structure des DataFrames poissons
print("poissons_indicateurs :")
print(indic.info())
print("\npoissons_observations :")
print(obs.info())
print("\npoissons_operations :")
print(ops.info())
print("\npoissons_stations :")
print(stats.info())
```

## On regarde les métadonnées communes au 4 tables (on va identifier les jointures primordiales et supprimer les données facultatives/inutiles pour notre étude)

```python
# Rappel : tu as déjà
# indic, obs, ops, stats

colonnes_indic = set(indic.columns)
colonnes_obs = set(obs.columns)
colonnes_ops = set(ops.columns)
colonnes_stats = set(stats.columns)

colonnes_communes = colonnes_indic & colonnes_obs & colonnes_ops & colonnes_stats

print("Colonnes communes aux 4 fichiers Poissons :")
for col in sorted(colonnes_communes):
    print(col)
```
